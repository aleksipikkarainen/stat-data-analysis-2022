```python
from scipy import stats
import csv
import pandas as pd

#1.
dataA = (34,23,52,47,34)
dataB = (48,28,33,46,41,35)
dataC = (34,53,51,35,52,32)

#Test for normality first for better test assessing
print(stats.shapiro(dataA)[1])
# p=0.67, normal

print(stats.shapiro(dataB)[1])
# p=0.74, normal

print(stats.shapiro(dataC)[1])
# p=0.03, not normal

#A&B, unpaired t-test, both normal
stats.ttest_ind(dataA,dataB)[1]
# p=0.93, not statistically significant

#A&C, unpaired Mann-Whitney U-test, C not normal
stats.mannwhitneyu(dataA,dataC,use_continuity=True,alternative='two-sided')[1]
# p=0.46, not statistically significant

#B&C, paired Wilcoxon test, C not normal
stats.wilcoxon(dataB,dataC,correction=True)[1]
# p=0.56, not statistically significant

#note that the sample size for the last two was too small
#for proper results

#2
#a) two-tailed Wilcoxon rank-sum test
#b) --||--
#c) One-way ANOVA with post hoc Bonferroni correction
#d) Two-tailed Studentâ€™s t test
#I think they used paired t-tests, all the values are paired
#And have a bit of SD, so you could compare their
#Normality distributions

#3
testD = (5.6,3.1,8.7,4.5,6.7,4.5)
retestD = (6.1,5.8,8.5,5.3,7.2,5.1)
testE = (4.5,3.9,7.1,4.3,6.9,8.2,7.6)
retestE = (4.9,4.7,7.8,4.8,7.5,7.8,8.1)

#Normality tests:
stats.shapiro(testD)[1]
# p=0.8, normal

stats.shapiro(retestD)[1]
# p=0.37, normal

stats.shapiro(testE)[1]
# p=0.16, normal

stats.shapiro(retestE)[1]
# p=0.014, not normal

#a)
stats.ttest_rel(testD,retestD)[1]
# p=0.097, not statistically significant

#b)
stats.wilcoxon(testE,retestE,correction=True)[1]
# p=0.03, statistically significant

#c)
stats.ttest_ind(testD,testE)[1]
# p=0.6, not statiscally significant

#d)
stats.mannwhitneyu(retestD,retestE,use_continuity=True,alternative="two-sided")[1]
# p= 0.94, not statistically significant

#4

#a)
stats.pearsonr(testE,retestE)
# r=0.97, p=0.00014, statistically significant 

#b)
stats.spearmanr(testE,retestE)
# r=0.93, p=0.001, statistically significant

#c)

#testE = (3,1,5,2,4,7,6)
#retestE = (3,1,5.5,2,4,5.5,6)
rankedtestE = (3,1,5,2,4,7,6)
rankedretestE = (3,1,5.5,2,4,5.5,6)

#d)
stats.pearsonr(rankedtestE,rankedretestE)
# r=0.95, p=0.0006, statistically significant

#5
df = pd.read_csv("simulated_data_2_5.csv")
colF = df["F"].to_numpy()
colG = df["G"].to_numpy()

#Check for normality
stats.shapiro(colF)[1]
# p=0.16, normal

stats.shapiro(colG)[1]
# p=0.45, normal

#a)
stats.ttest_rel(colF,colG)
# p=0.37, not statistically significant

stats.pearsonr(colF,colG)
# c = 0.03, p=0.75, little to no correlation and not statistically significant

#b)
colH = df["H"].to_numpy()

#Check for normality
stats.shapiro(colH)[1]
# p=0.94, normal

stats.ttest_rel(colF,colH)
#  p=0.11,  not statistically significant

stats.pearsonr(colF,colH)
# correlation=-0.14, p=0.15, negative correlation and not stastically significant

#c)
colI = df["I"].to_numpy()

#Check for normality
stats.shapiro(colI)
# p = 0.001, not normally distributed

stats.wilcoxon(colF,colI,correction=True)[1]
# p = 0.00008, statistically significant

stats.spearmanr(colF,colI)
# corr = 0.09, p=0.33, little to no correlation and not statistically significant



```
