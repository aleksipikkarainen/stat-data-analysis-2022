```python

import pandas as pd
import matplotlib.pyplot as plt
import sklearn.cluster as SC
import sklearn.decomposition as SD
import sklearn.preprocessing as SP
iris = pd.read_csv("iris_csv.csv")
```

```python
#1
iris.plot(kind ="scatter",
          x ='sepallength',
          y ='sepalwidth')
plt.grid()
```

```python
scatter = pd.DataFrame(iris, columns=["sepallength","sepalwidth","petallength","petalwidth"])
pd.plotting.scatter_matrix(scatter)
```

```python

df_original = iris[["sepallength","sepalwidth"]]
df_standardised = SP.StandardScaler().fit(df_original).transform(df_original)
pca = SD.PCA().fit(df_standardised)
df_pca = pd.DataFrame(pca.transform(df_standardised))
iris_extended = pd.concat([iris, df_pca], axis=1)
iris_extended.plot.scatter(0, 1, c='class', colormap='brg', figsize=(12,10))
```

```python
pd.DataFrame(pca.components_, columns=df_original.columns)
```

```python
pairs = pd.DataFrame(iris, columns=["sepallength","sepalwidth","petallength","petalwidth"])
corr = pairs.corr(method="spearman")
pd.plotting.scatter_matrix(corr)
```

```python
pairs = pd.DataFrame(iris_extended, columns=["sepallength","sepalwidth","petallength","petalwidth"])
corr = pairs.corr(method="spearman")
pd.plotting.scatter_matrix(corr)
```

```python
# create k-means model with three clusters
kmeans = SC.KMeans(n_clusters=3).fit(df_standardised)
# get memberships of data points in clusters
predictions = kmeans.predict(df_standardised)
# add memberships back to original data frame
iris_extended['cluster'] = pd.Categorical(predictions)

# visualise the first two PCA components and
# colour data points by cluster membership
iris_extended.plot.scatter(0, 1, c='cluster', colormap='brg', figsize=(12,10))
```
